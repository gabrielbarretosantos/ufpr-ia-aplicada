```python
import pandas as pd
import numpy as np
from kmodes.kmodes import KModes
import warnings
from IPython.display import display
warnings.filterwarnings("ignore")
pd.set_option('display.float_format', lambda x: '%.3f' % x)
SEED_VALUE = 202407
vehicles = pd.read_csv("data/4 - Veiculos - Dados.csv")
```

```python

vehicles.drop(columns=['a'], inplace=True)
np.random.seed(SEED_VALUE)
km = KModes(n_clusters=10, init='Huang', n_init=10, verbose=1, random_state=SEED_VALUE)
clusters = km.fit_predict(vehicles)
vehicles['cluster'] = clusters
print(vehicles.head())
display(vehicles)
```
